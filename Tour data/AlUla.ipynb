```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataframe
messy_df = pd.read_csv('AlUla.csv')
```

```python
# take a look of the first 5 rows in the data
messy_df.head()
```

```python
# check the datatype & non-null count
messy_df.info()
```

```python
# making a cop of the data to clean it
df = messy_df.copy()
```

```python
# dropping the unnecessary columns
df.drop(['cctrW', 'FmrIP href', 'dugSS src', 'cIVmB', 'ySnDE', 'dXxty', 'ccVGl', 'WlYyy 4'], axis=1 , inplace=True)
```

```python
# renaming the columns
df.rename(columns={'iPqaD':'Tour By', 'bUshh':'Tour Name', 'iPqaD href':'Link', 'WlYyy':'Tour Type', 'WlYyy 2':'Tour Duration', 'WlYyy 6':'Cancellation', 'WlYyy 3':'Tour Price', 'ccVGl 2':'Price Per'}, inplace=True)
```

```python
# adding new columns
df['City'] = 'Al Ula'
df['Country'] = 'Saudi Arabia'
```

```python
# check the total duplicated rows
sum(df.duplicated())
```

```python
df.columns
```

```python
# Reorfer the columns
df = df[['Tour Name', 'Tour Type', 'Tour Duration', 'Tour Price', 'Price Per', 'Cancellation','Tour By', 'City', 'Country', 'Link']]
```

```python
#checking for total null values
df.isnull().sum() 
```

```python
# reset the index
df.reset_index(drop=True)
```

```python
df.info()
```

**For the Cleaning:**

- Tour Type:
    - Remove these values (16 places sorted by traveler favorites , Frequently Asked Questions about AlUla)
- Tour Duration:
    - Duration in (hours), 
    - some value has + sign with the number, 
    - Remove this value (Which experiences are best for tours in AlUla?)
- Tour Price:
    - Remove this value(What are the best places for tours in AlUla? )
    - Remove $, 
    - convert to float type
- Tour By:
    - Remove this value (Clear all filters)
    

```python
# Store the clean dataset
df.to_csv('AlUlaclean.csv')
```
