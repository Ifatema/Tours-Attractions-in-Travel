```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataframe
messy_df = pd.read_csv('Antalya.csv')
```

```python
# take a look of the first 5 rows in the data
messy_df.head()
```

```python
# check the datatype & non-null count
messy_df.info()
```

```python
# making a cop of the data to clean it
df = messy_df.copy()
```

```python
# dropping the unnecessary columns
df.drop(['iPqaD href', 'FmrIP href', 'dugSS src', 'WlYyy 2', 'ySnDE', 'ccVGl', 'cIVmB', 'WlYyy 9', 'WlYyy 5'], axis=1 , inplace=True)
```

```python
# renaming the columns
df.rename(columns={'iPqaD':'Tour By', 'bUshh':'Tour Name', 'iPqaD href 2':'Link', 'WlYyy 3':'Tour Type', 'WlYyy 4':'Tour Duration', 'WlYyy 6':'Cancellation', 'WlYyy 8':'Tour Price', 'ccVGl 2':'Price Per'}, inplace=True)
```

```python
# adding new columns
df['City'] = 'Antalya'
df['Country'] = 'Turkey'
```

```python
# check the total duplicated rows
sum(df.duplicated())
```

```python
# drop the duplicated rows
df.drop_duplicates()
```

```python
df.columns
```

```python
# Reorfer the columns
df = df[['Tour Name', 'Tour Type', 'Tour Duration', 'Tour Price', 'Price Per', 'Cancellation','Tour By', 'City', 'Country', 'Link']]
```

```python
#checking for total null values
df.isnull().sum() 
```

```python
# reset the index
df.reset_index(drop=True)
```

```python
df.info()
```
