```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataframe
messy_df = pd.read_csv('Cape Town.csv')
```

```python
messy_df.columns
```

```python
# take a look of the first 5 rows in the data
messy_df.head()
```

```python
# check the datatype & non-null count
messy_df.info()
```

```python
# making a cop of the data to clean it
df = messy_df.copy()
```

```python
# dropping the unnecessary columns
df.drop(['WlYyy','cctrW','iPqaD href', 'FmrIP href','cIVmB','ySnDE', 'WlYyy 5', 'cctrW 2', 'ccVGl', 'WlYyy 7', 'dXxty'], axis=1 , inplace=True)
```

```python
# renaming the columns
df.rename(columns={ 'dugSS src':'Image','iPqaD':'Tour By', 'bUshh':'Tour Name', 'iPqaD href 2':'Link', 'WlYyy 2':'Tour Type', 'WlYyy 3':'Tour Duration', 'WlYyy 4':'Cancellation', 'WlYyy 6':'Tour Price', 'ccVGl 2':'Price Per'}, inplace=True)
```

```python
# adding new columns
df['City'] = 'Cape Town'
df['Country'] = 'South Africa'
```

```python
# check the total duplicated rows
sum(df.duplicated())
```

```python
# drop the duplicated rows
df.drop_duplicates()
```

```python
df.columns
```

```python
# Reorfer the columns
df = df[['Image','Tour Name', 'Tour Type', 'Tour Duration', 'Tour Price', 'Price Per', 'Cancellation','Tour By', 'City', 'Country', 'Link']]
```

```python
#checking for total null values
df.isnull().sum() 
```

```python
# reset the index
df.reset_index(drop=True)
```

```python
df.info()
```

```python
# Store the clean dataset
df.to_csv('Cape Townclean.csv')
```
